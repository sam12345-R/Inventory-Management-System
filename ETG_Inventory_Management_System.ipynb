#Data Creation

>In this section, I just want to show you how I created my data via coding instead of doing that manually.. The project starts directly after this section (IMS Features).

```python
''' The data (products) is in JSON format where the key is an automated ID (starts from 1000) 
    and the value (product) is also in JSON format where the keys are 8: prod_name - manufacturer - category - quantity -
    price - review - status - and created_at.
      - quantity and price should be positive
      - review should be between 1 and 5
      - status means whether the product is purchased or no. By default, contains "-" means not purchased so far.
      '''
products = {}
product = {}
```

```python
import time
```

```python
while True:
  print('---------------------------------------')
  product['prod_name'] = input('Enter product Name: ')
  product['manufacturer'] = input('Enter product Manufacturer: ')
  product['category'] = input('Enter product Category: ')
  product['quantity'] = abs(int(input('Enter product Quantity: ')))
  product['price'] = abs(float(input('Enter product Price: ')))
  product['review'] = abs(int(input('Enter product Review (1 - 5): ')))
  while (product['review'] < 1 or product['review'] > 5):
    product['review'] = abs(int(input('Invalid review, please make sure to enter a review between 1 and 5: ')))
  product['status'] = '-'
  product['created_at'] = time.strftime('%d/%m/%Y %H:%M:%S')
  products[len(products) + 1000] = product
  product = {}
  stop = input('Would you like to continue adding products? (type "n" if you want to stop): ')
  if (stop[0].lower() == 'n'):
    break
```

# IMS Features
Now I will be working on some of IMS features like showing all products, adding new product, purchasing a product and showing report.

```python
# I'll be using this dictionary to save all tasks like add, update purchase, delete product for reporting purpose
# the key will be the datetime of the task and the value would be a sentence.
# You will understand what I mean when you reach the end of the project  
import time
report = {}
```

```python
# I have to read the Records.json file to work on it
with open('Records.json', 'r') as f:
  txt = f.read()  
```

```python
import json
```

```python
# To facilitate my work, I need to transfer the content of the file (string format) into JSON format
prods = json.loads(txt)
prods
```

-------------- Displaying Products --------------
 
I'll be using the tabulate library for this purpose 

```python
from tabulate import tabulate
```

```python
# I'll display the products depending on each category. 
# As long as tabulate needs two separate values (Data and Headers), I'll extract these values into different arrays
```

```python
# Extracting all the unique categories 

categories = []
for product in prods.values():
  if product['category'] not in categories:
    categories.append(product['category'])

categories
```

```python
# The tabulate method needs the data to be an array of tuples a header array for columns titles 
data = []
headers = ['ID', 'Name', 'Manufacturer', 'Quantity', 'Price', 'Review', 'Status', 'Created At']
for category in categories:
  for id, product in prods.items():
    if product['category'] == category:
      # data should be in tuple format
      data.append(tuple((id, product['prod_name'], product['manufacturer'], product['quantity'], '$' + str(product['price']), '* ' * product['review'], product['status'], product['created_at'])))
  print('\t\t\t >>>>>>>>>>>>>>>>>>>>>>>>', category, '<<<<<<<<<<<<<<<<<<<<<<<<\n\n')
  print(tabulate(data, headers=headers, tablefmt='orgtbl', stralign='center', numalign='center'), '\n\n')
  data = [] # If I didn't empty the data array, there would be a lot of duplications
```

I'll be working on the 'add new product' feature 

```python
# It's same way that I used to create Records.txt. But, The only different things are :
#                                                       - one product can be added at a time 
#                                                       - I noticed that id is now of type string (previously it was integer)
product = {}
print('>>>>>>>>>>>>>>>>>>>>>>>> Add Product <<<<<<<<<<<<<<<<<<<<<<<<')
product['prod_name'] = input('Enter product Name: ')
product['manufacturer'] = input('Enter product Manufacturer: ')
product['category'] = input('Enter product Category: ')
product['quantity'] = abs(int(input('Enter product Quantity: ')))
product['price'] = abs(float(input('Enter product Price: ')))
while (product['review'] < 1 or product['review'] > 5):
  product['review'] = abs(int(input('Invalid review, please make sure to enter a review between 1 and 5: ')))
product['review'] = abs(int(input('Enter product Review (1 - 5): ')))
product['status'] = '-'
product['created_at'] = time.strftime('%d/%m/%Y %H:%M:%S')
prods[str(len(prods) + 1000)] = product

report[time.strftime('%d/%m/%Y %H:%M:%S')] = 'A product of name {} has been added by an Admin'.format(product['prod_name'])
```

-------------- Purchasing Product --------------
 
It's time to working on the 'purchase product' feature 

```python
''' I'm creating a new data that contains all the possible informations of purchases. the data (sales) is in JSON format.
    (starts from 1). The key is an automated ID and the value (invoice) is a structured format (JSON and also called NoSQL)
    The invoice data contains 6 keys: username - payment_method - products - amount - bill_to and ship_to if payment_method is
    in-place else we have card_num and card_code. products key is a 2-dimensional array where each sub-array contains
    product ID, product name, purchased quantity and purchased time. The amount key is the sum of all purchases price.'''
sales = {}
invoice = {}
purchased_prods = []
data = []
headers = ['ID', 'Qty', 'Product Name', 'Category', 'Unit Price', 'Amount', 'purhased_at']
```

```python
# First, I will prompt the user to enter the desired product (id and quantity)
user_name = input('Please, enter your name: ')
# Asking the user about its payment method (it should be either online or in-place)
payment_method = input('Would you like to pay online on in-place? (type "online" / "in-place")').lower()
while (payment_method not in ['online', 'in-place']):
  payment_method = input('Invalid payment method. Type a valid one: ').lower()
if (payment_method == 'in-place'):
  bill_to = input('Please, enter the "bill to" address: ')
  ship_to = input('Please, enter the "ship to" address: ')
  invoice['bill_to'] = bill_to
  invoice['ship_to'] = ship_to
else:
  invoice['card_num'] = input('Please, enter your card number: ')
  invoice['card_code'] = input('Please, enter your card code: ')

# Taking some details from the user for the purchasing process 
amount = 0
while (True):
  id_by_user = input('Please, enter the ID of the product you want to purchase: ')
  while (id_by_user not in prods.keys()):
    id_by_user = input('Oops, the ID you entered is not shortlisted here.. Please, re-order again!!')
  prods[id_by_user]['status'] = 'Purhased'
  prod_qte = abs(int(input('Please, enter the quantity you tend to purchase for the product No.' + id_by_user + ': ')))
  while (prod_qte > prods[id_by_user]['quantity']):
    print('Sorry, your order is higher than what we have in the stock.')
    print('You want to purchase', prod_qte, 'but we only have', prods[id_by_user]['quantity'], 'in the stock.')
    prod_qte = abs(int(input('Please, re-enter a valid quantity for the product No.' + id_by_user + ': ')))
  prods[id_by_user]['quantity'] -= prod_qte
  purchased_prods.append([id_by_user, prods[id_by_user]['prod_name'], prod_qte, time.strftime('%d/%m/%Y %H:%M:%S')])
  amount += (prods[id_by_user]['price'] * prod_qte)
  stop = input('Would you like to order more products? (type "n" if you want to stop): ')
  if (stop[0].lower() == 'n'):
    break
# I am assuming that there's a tax on the sales
tax = amount * 0.05 # Sale tax: 5%

# Stocking the informations in the invoice dictionary then save them in the sales dictionary
invoice['username'] = user_name
invoice['payment_method'] = payment_method
invoice['products'] = purchased_prods
invoice['amount'] = amount 
sales[len(sales) + 1] = invoice

report[time.strftime('%d/%m/%Y %H:%M:%S')] = '{} has purchased {} products under amount {} (without tax) and decided to pay {}'.format(
                            user_name, len(invoice['products']), invoice['amount'], invoice['payment_method'])
```

```python
import datetime
```

```python
# Displaying the invoice report using the tabulate package as well 

print('>'*25,'Invoice Report for', invoice['username'], '<'*25)
if (invoice['payment_method'] == 'online'):
  print('\nVerifying your card code...', end=' ')
  time.sleep(3)
  print('done.')
  print('Verifying your card number...', end=' ')
  time.sleep(3)
  print('done.')
  sec = 10
  print('Please wait.. the transaction process will ends up after: ')
  while (sec >= 0):
    print(str(sec) + '..', end=' ')
    sec -= 1
    time.sleep(1)
  print(' Transaction ended successfully.')
  print('\n-------------------------------------------------------------------------\n')
  print('Invoice #' + str(len(sales)))
  print('User Name: ', invoice['username'])
  print('Payment Method: ', invoice['payment_method'])
  print('Invoice Date: ', time.strftime('%d/%m/%Y %H:%M:%S'))
  print('Due Date: ', (datetime.date.today() + datetime.timedelta(days=15)), '\n')
  pr = invoice['products']
  for p in range(len(pr)):
    data.append(tuple((pr[p][0], pr[p][2], pr[p][1], prods[pr[p][0]]['category'], '$'+str(prods[pr[p][0]]['price']), 
                       '$'+str(prods[pr[p][0]]['price']*pr[p][2]), pr[p][-1])))
  print(tabulate(data, headers=headers, tablefmt='orgtbl', stralign='center', numalign='center'), '\n\n')
  data = []
  print('\nSubTotal =\t', ('$' + str(round(invoice['amount'], 2))))
  print('Tax (5%) =  \t', ('$' + str(round(tax, 2))))
  print('Total    =  \t', ('$' + str(round(invoice['amount'] + tax, 2))))
  print('\nThanks for trusting us ' + invoice['username'] + '.. Happy Collaboration!')
else:
  sec = 10
  print('\nPlease wait.. the preparation process will ends up after: ')
  while (sec >= 0):
    print(str(sec) + '..', end=' ')
    sec -= 1
    time.sleep(1)
  print('Preparation ended successfully.')
  print('\n-------------------------------------------------------------------------\n')
  print('Invoice #' + str(len(sales)))
  print('User Name: ', invoice['username'])
  print('Bill To: ', invoice['bill_to'])
  print('Ship To: ', invoice['ship_to'])
  print('Payment Method: ', invoice['payment_method'])
  print('Invoice Date: ', time.strftime('%d/%m/%Y %H:%M:%S'))
  print('Due Date: ', (datetime.date.today() + datetime.timedelta(days=15)), '\n')
  pr = invoice['products']
  for p in range(len(pr)):
    data.append(tuple((pr[p][0], pr[p][2], pr[p][1], prods[pr[p][0]]['category'], '$'+str(prods[pr[p][0]]['price']),
                       '$'+str(prods[pr[p][0]]['price']*pr[p][2]), pr[p][-1])))
  print(tabulate(data, headers=headers, tablefmt='orgtbl', stralign='center', numalign='center'), '\n')
  print('\nSubTotal =\t', ('$' + str(round(invoice['amount'], 2))))
  print('Tax (5%) =  \t', ('$' + str(round(tax, 2))))
  print('Total    =  \t', ('$' + str(round(invoice['amount'] + tax, 2))))
  print('\nThanks for trusting us ' + invoice['username'] + '.. Happy Collaboration!')
  data = []
invoice = {} # I should empty the data array and the invoice dictionary to avoid duplication
```

-------------- Deleting Product --------------
 
I'll be working now on the 'delete product' feature 

```python
# In this feature, I'll just prompt the admin to enter the product ID for deletion
print('>>>>>>>>>>>>>>>>>>>>>>>> Delete Product <<<<<<<<<<<<<<<<<<<<<<<<\n')
del_id = input('Hi Admin, please, enter the ID of the product you seek to delete: ')
while (del_id not in prods.keys()):
  del_id = input('Oops, the ID you entered is not shortlisted here.. Please, enter an available one!!')

del prods[del_id]
print('Product No.' + del_id + ' has been deleted successfully!')

report[time.strftime('%d/%m/%Y %H:%M:%S')] = 'A product of ID #{} has been deleted by an Admin'.format(del_id)
```

-------------- Update Product --------------
 
I'll be working now on the 'update product' feature 

```python
# In this feature, I'll just prompt the admin to enter the product ID for updating
print('>>>>>>>>>>>>>>>>>>>>>>>> Update Product <<<<<<<<<<<<<<<<<<<<<<<<\n')
update_id = input('Hi Admin, please, enter the ID of the product you seek to update: ')
while (update_id not in prods.keys()):
  update_id = input('Oops, the ID you entered is not shortlisted here.. Please, enter an available one!!')

while True:
  ans = input('What do you want to update among these: ["prod_name", "manufacturer", "category", "quantity", "price", "review"] ')
  while (ans not in prods[update_id].keys()):
    ans = input('Please, choose a valid one: ["prod_name", "manufacturer", "category", "quantity", "price", "review"] ')
  if (ans == 'prod_name'):
    prods[update_id]['prod_name'] = input('Enter product Name: ')
  elif (ans == 'manufacturer'):
    prods[update_id]['manufacturer'] = input('Enter product Manufacturer: ')
  elif (ans == 'category'):
    prods[update_id]['category'] = input('Enter product Category: ')
  elif (ans == 'quantity'):
    prods[update_id]['quantity'] = abs(int(input('Enter product Quantity: ')))
  elif (ans == 'price'):
    prods[update_id]['price'] = abs(float(input('Enter product Price: ')))
  elif (ans == 'review'):
    while (prods[update_id]['review'] < 1 or prods[update_id]['review'] > 5):
      prods[update_id]['review'] = abs(int(input('Invalid review, please make sure to enter a review between 1 and 5: ')))
    prods[update_id]['review'] = abs(int(input('Enter product Review (1 - 5): ')))
  print('Product No.' + update_id + ' has been updated successfully!')
  report[time.strftime('%d/%m/%Y %H:%M:%S')] = 'An Admin has changed the {} of product #{}'.format(ans, update_id)
  stop = input('Would you like to continue updating products? (type "n" if you want to stop): ')
  if (stop[0].lower() == 'n'):
    break
```
